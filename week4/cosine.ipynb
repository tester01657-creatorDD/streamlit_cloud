# Embedding

```python
import ollama

# API 키 불필요 — 로컬 서버가 대신함
# 사전 준비: 터미널에서 ollama pull nomic-embed-text

sentences = [
    "오늘 서울은 맑고 따뜻한 봄 날씨입니다",
    "내일도 화창하고 기온이 올라갈 예정입니다",
    "파이썬에서 리스트를 정렬하려면 sort 메서드를 사용합니다",
]

vectors = []
for s in sentences:
    result = ollama.embed(
        model="nomic-embed-text",
        input=s
    )
    vectors.append(result['embeddings'][0])
    print(f"'{s}' → 앞 5개: {result['embeddings'][0][:5]}")
    
```

# 직접 구현 (원리 이해용)

```python
import math

def cosine_similarity(a, b):
    # 내적: 같은 위치의 숫자끼리 곱해서 합산
    dot_product = sum(x * y for x, y in zip(a, b))
    
    # 각 벡터의 크기 (길이)
    norm_a = math.sqrt(sum(x ** 2 for x in a))
    norm_b = math.sqrt(sum(x ** 2 for x in b))
    
    # 유사도 = 내적 / (크기 × 크기)
    return dot_product / (norm_a * norm_b)

# 테스트
sim_12 = cosine_similarity(vectors[0], vectors[1])
sim_13 = cosine_similarity(vectors[0], vectors[2])

print(f"'오늘 서울은 맑고 따뜻한 봄 날씨입니다' vs '내일도 화창하고 기온이 올라갈 예정입니다': {sim_12:.4f}")
print(f"'오늘 서울은 맑고 따뜻한 봄 날씨입니다' vs '파이썬에서 리스트를 정렬하려면 sort 메서드를 사용합니다': {sim_13:.4f}")
```

# numpy로 간결하게

```python
import numpy as np

def cosine_sim_np(a, b):
    a, b = np.array(a), np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# 결과는 동일
print(f"numpy 버전: {cosine_sim_np(vectors[0], vectors[1]):.4f}")
print(f"numpy 버전: {cosine_sim_np(vectors[0], vectors[2]):.4f}")
```
